# Preprocessing Data

## Data Cleaning

```
# Menangani dan mengganti missing value
# Pada dataset Zomato, nilai kosong sering ditulis sebagai 'NaN' (dengan spasi)
df.replace('Nan', np.nan, inplace=True)
df.dropna(inplace=True)
```

```
# Memperbaiki tipe data
# Mengubah usia dan rating menjadi numerik
df['Delivery_person_Age'] = pd.to_numeric(df['Delivery_person_Age'], errors='coerce')
df['Delivery_person_Ratings'] = pd.to_numeric(df['Delivery_person_Ratings'], errors='coerce')
```

```
# Menghapus teks '(min)' dari kolom target agar menjadi angka
df['Time_taken (min)'] = df['Time_taken (min)'].astype(float)
```

```
# Filtering koordinat tidak valid
# Menghapus baris yang koordinatnya 0 karena tidak mungkin di tengah laut (0,0)
df = df[df['Restaurant_latitude'] != 0]
```

## Data Manipulating

```
from geopy.distance import geodesic

def feature_engineering_process(df):
    # A. Menghitung Jarak (Distance Feature)
    # Mengubah 4 kolom koordinat menjadi 1 kolom Jarak dalam KM
    def get_distance(row):
        start = (row['Restaurant_latitude'], row['Restaurant_longitude'])
        stop = (row['Delivery_location_latitude'], row['Delivery_location_longitude'])
        return geodesic(start, stop).km

    df['distance_km'] = df.apply(get_distance, axis=1)

    # B. Ekstraksi Waktu (Time Features)
    # Membuat kategori waktu: Pagi, Siang, Sore, Malam dari jam pemesanan
    # (Asumsi kolom 'Time_Ordered' sudah dibersihkan)
    df['order_hour'] = pd.to_datetime(df['Time_Orderd'], format='%H:%M', errors='coerce').dt.hour # Corrected Time_Orderd column name here

    # C. Klasifikasi Kecepatan (Efficiency Feature)
    # Menghitung seberapa cepat kurir menempuh 1 KM
    df['min_per_km'] = df['Time_taken (min)'] / df['distance_km'] # Corrected column name here

    # D. Binning (Kategorisasi)
    # Mengubah usia menjadi kelompok umur agar lebih mudah dianalisis
    bins = [15, 25, 35, 50]
    labels = ['Young', 'Adult', 'Senior']
    df['age_group'] = pd.cut(df['Delivery_person_Age'], bins=bins, labels=labels)

    return df

# Eksekusi
df_final = feature_engineering_process(df)
```

```
def check_outliers(df):
    plt.figure(figsize=(12, 5))

    # Visualisasi Jarak
    plt.subplot(1, 2, 1)
    sns.boxplot(y=df['distance_km'])
    plt.title('Outlier Jarak (km)')

    # Visualisasi Waktu
    plt.subplot(1, 2, 2)
    sns.boxplot(y=df['Time_taken (min)'], color='orange')
    plt.title('Outlier Waktu (min)')

    plt.tight_layout()
    plt.show()

check_outliers(df_final)
```

Terlihat outlier dari jarak sangat jauh dari batas normal

```
def remove_outliers_iqr(df, columns):
    df_clean = df.copy()

    for col in columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Filter data
        initial_count = len(df_clean)
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
        final_count = len(df_clean)

        print(f"Kolom {col}: Menghapus {initial_count - final_count} baris outlier.")

    return df_clean

# Eksekusi pembersihan pada kolom Jarak dan Waktu
df_filtered = remove_outliers_iqr(df_final, ['distance_km', 'Time_taken (min)'])
```

```
# Menangani Outlier pada Jarak dan Waktu
# Menggunakan IQR untuk membuang data yang tidak masuk akal (misal jarak 0km tapi waktu 100 menit)
q_low = df["distance_km"].quantile(0.01)
q_hi  = df["distance_km"].quantile(0.99)
df = df[(df["distance_km"] < q_hi) & (df["distance_km"] > q_low)]
```

```
# Menghapus data yang secara logika tidak mungkin
# Contoh: Jarak lebih dari 50km untuk pengiriman makanan dalam kota diragukan
df_filtered = df_filtered[df_filtered['distance_km'] < 50]

# Contoh: Waktu pengiriman kurang dari 5 menit (terlalu cepat, mungkin error input)
df_filtered = df_filtered[df_filtered['Time_taken (min)'] >= 5]

print(f"Data akhir setelah pembersihan outlier: {df_filtered.shape[0]} baris")
```

- Pengaruh Eksternal: Kondisi cuaca berkabut (Fog) dan badai (Stormy) secara konsisten meningkatkan waktu pengiriman. Hal ini menunjukkan perlunya margin waktu tambahan dalam estimasi pesanan saat cuaca buruk.

- Performa Armada: Motor memang lincah, namun pada tingkat kemacetan tertinggi, semua jenis kendaraan mengalami hambatan yang serupa. Efisiensi kendaraan sangat bergantung pada kepadatan jalan daripada jenis mesinnya.

- Segmentasi Usia: Kelompok usia kurir menunjukkan performa yang relatif stabil, namun terdapat kecenderungan kelompok tertentu memiliki konsistensi waktu yang lebih baik pada jarak menengah.

- Titik Lemah Operasional: Telah teridentifikasi 5 koordinat restoran dengan waktu tunggu/pengiriman tertinggi yang menjadi bottleneck utama dalam jaringan distribusi saat ini.
